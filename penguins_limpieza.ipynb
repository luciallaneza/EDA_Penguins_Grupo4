### Tabla de decisiones de calidad del dato

Completad esta tabla durante la limpieza y actualizadla si cambian vuestras decisiones.

| Problema detectado | Columna afectada | Decision tomada | Justificacion | Impacto esperado |
|---|---|---|---|---|
| 2 Pingüinos con estádisticas nulas | Culmen Length (mm)    Culmen Depth (mm)    Flipper Length (mm)    Body Mass (g)| Eliminación de datos | No son relevantes | Ninguno |

```python
#Eliminar filas

df_limpio = df.dropna(subset=["Culmen Length (mm)"])
```

```python
#Eliminar filas

df_limpio = df.dropna(subset=["Delta 15 N (o/oo)"])
```

```python
print('Filas originales:', len(df))
print('Filas tras limpieza:', len(df_limpio))
print('\nNulos tras limpieza:')
print(df_limpio.isna().sum())
```

```python
countsex = df['Sex'].value_counts()

print(countsex)
# Acceso específico:
print(f"Male: {countsex.get('macho', 0)}")
print(f"Female: {countsex.get('hembra', 0)}")
```

```python
#Otra forma de verlo

male = (df["Sex"] == "MALE").sum()

female = (df["Sex"] == "FEMALE").sum()

print(f"Male:  {male}")
print(f"Female:  {female}")
```

```python
#Rellenamos los nulos de Sex con "indefinido"


df_limpio['Sex'] = df_limpio['Sex'].fillna('indefinido')

df_limpio
```

```python
#Comprobación

df_limpio[df_limpio["Individual ID"] == "N5A2"]
```

```python
df_limpio.drop(columns=["Region"], inplace=True)
df_limpio.drop(columns=["Clutch Completion"], inplace=True)
df_limpio.drop(columns=["Comments"], inplace=True)
```
