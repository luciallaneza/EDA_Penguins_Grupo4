# 🧱 Limitaciones del análisis

#### 🤷🏻‍♀️ Qué no se puede concluir con estos datos:

No podemos concluir diferencias respecto a su comportamiento (migración, reproducción y anidación).
<br>

#### ⚖ Qué sesgos o desequilibrios condicionan la interpretación:

No podemos hacer seguimiento real a lo largo de los años de los individuos ya que se reutilizan los IDs tanto para especies diferentes, sexo o ubicación. Solo sirve para el estudio en un periodo de tiempo limitado. 
Encontramos **errores sistemáticos** ya que a lo largo de los años se mantiene una identificación errónea de los pingüinos y las muestras del proyecto. 

Además, no podemos hacer una interpretación de ciertos datos sin tener conocimiento técnico (ej. columnas Delta) para evitar especulaciones sobre biología marina.


#### 📋 Qué información faltaría para responder mejor al cliente:

Una nomenclatura/codificación alfanumérica más optimizada para conseguir una trazabilidad más sencilla y accesible en el estudio y evitar caer en el sesgo de información.
