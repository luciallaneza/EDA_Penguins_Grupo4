# 🎯 Conclusiones y recomendaciones

## Primer hallazgo


#### 1. Evidencia observada
- Observamos un patrón en los IDs individuales, donde aunque se pierdan los datos o aparezcan como nulos, en su gran mayoría hacen referencia a una pareja (macho/hembra).

#### 2. Interpretación

- El estudio se enfoca en parejas de pingüinos.

#### 3. Implicación para el cliente

- Al tratarse de un dato muy relevante para el estudio, el cliente debe asegurar la calidad de este dato. 

#### 4. Recomendación concreta

- Asignar un número en el individual_ID (1 o 2) dependiendo de si se trata de un macho (por ej. 1) o una hembra (por ej. 2) para mantener la trazabilidad. 

## Segundo hallazgo

#### 1. Evidencia observada 

- Existe una gran diferencia en la cantidad de datos analizados respecto a las especies con las que tratamos, esto se ve reflejado en el análisis univariado en primer lugar analizando la cantidad de pingüinos por especie, siendo la especie Chinstrap la que menos muestras aporta (el 20%) en comparación a Adelie y Gentoo que suponen más del doble en ambos casos (casi el 40% cada uno)

#### 2. Interpretación

- El dataset no está equilibrado para hacer un análisis comparativo por especies. 

#### 3. Implicación para el cliente

- Para el cliente no va a aportar valor real en las comparaciones de muestras. 

#### 4. Recomendación concreta

- Recomendamos reconsiderar el muestreo de cada especie o que se especifique mejor en el análisis si estos datos son relevantes para comparaciones futuras. 



## Tercer hallazgo
#### 1. Evidencia observada 

- Basándonos en los estudios por islas vemos que la toma de muestras se encuentra concentrada en las islas Biscoe y Dream siendo la isla Torgersen significativamente menor y suponiendo un 13%.

#### 2. Interpretación

- Dataset escaso para hacer una distribución real (sin caer en suposiciones de tamaño de la isla o comportamiento de especies)

#### 3. Implicación para el cliente

- Debido a esta evidencia el análisis de los datos puede estar sesgado.

#### 4. Recomendación concreta

- Mejorar la calidad del dato para que las comparativas sean más fieles y si al cliente podría interesarle la presencia/ausencia de los pingüinos entre las islas en época de anidación e incubación. 

## Cuarto hallazgo
#### 1. Evidencia observada 

- Las especies Adelie y Chinstrap tienen características similares (profundidad del pico y longitud de la aleta) y a veces son difíciles de separar para interpretar perfiles biológicos.

#### 2. Interpretación

- Al tener datos similares las comparaciones no son relevantes en estas especies. 

#### 3. Implicación para el cliente

- Si el objetivo es obtener diferencias observables entre estas especies será necesario aportar otros datos. 

#### 4. Recomendación concreta

- Recoger otro tipo de datos más significativos y que los diferencien entre ellos, ampliando la toma de muestras y creando perfiles biológicos más fieles que ayuden a comparar la morfología de las especies. 


## Quinto hallazgo
#### 1. Evidencia observada 

- Pocos datos temporales (solo en el mes de noviembre y principios de diciembre) y de población en general para poder tener conocimientos de la población anual de las islas.


#### 2. Interpretación

- Podemos suponer que se trata de la toma de datos de la época concreta de mayor anidación y reproducción de las especies, pero no podemos concluir si aumentan o disminuyen ya que no conocemos la población total durante el año.

#### 3. Implicación para el cliente

- Tendría datos más fieles del aumento o decrecimiento de la población en comparación con el dataset que tenemos que solo abarca unas fechas concretas de cuándo han encontrado ese huevo 

#### 4. Recomendación concreta

- Mantener un seguimiento anual o estacional para ver la evolución de la distribución poblacional. El cliente tendría que aumentar el número de muestras dentro de la población por islas (crías, pingüinos adultos en epoca de reproducción o pingüinos más mayores).
